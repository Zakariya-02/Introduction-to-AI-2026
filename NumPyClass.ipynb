**Array indexing**

```python
students = np.array(["Irena", "Irfan", "Moksh", "Rudra", "Vinicius", "Mbappe"])
print(students[0])
print(students[-1])
```

**Array slicing: slicing means choosing multiple elements**

```python
print(students[1:4])
print(students[:3]) # prints first 3 elements
print(students[-2:]) # prints last 2 elements
print(students[::2]) # skips values
```

**Mathematical operations**

```python
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([7, 8, 13, 14, 5])
print(array1 + array2) #addition
print(array1 - array2) #subtraction
print(array1 * array2) #multiplaction
print(array1 / array2) #Division
print(array1*2) #multiplying by a number
```
